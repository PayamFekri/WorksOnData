### Face Recognition

```python
api_key = "acc_b910ac569da00e2"
api_secret = "8fd8cbd6a4455778525c679fe77ab732"

url_detection = "https://api.imagga.com/v2/faces/detection"
params1 = {
    'image_url': 'https://postimg.cc/CnRf0xJm'
}

response1 = requests.get(url_detection, params=params1, auth=HTTPBasicAuth(api_key, api_secret))
```

```python
response1.text , response1.status_code
```

```python
data1 = response1.json()
print("Detection result image1:", data1)
face_id1 = data1['result']['faces'][0]['face_id']
face_id1
```

```python
# مرحله 2: گرفتن face_id از تصویر دوم
params2 = {'image_url': 'https://example.com/face2.jpg'}

response2 = requests.get(url_detection, params=params2, auth=HTTPBasicAuth(api_key, api_secret))
data2 = response2.json()
print("Detection result image2:", data2)

face_id2 = data2['result']['faces'][0]['face_id']  # اولین چهره
```

```python
# مرحله 3: مقایسه دو چهره با Face Similarity
url_similarity = "https://api.imagga.com/v2/faces/similarity"
params_similarity = {
    'face_id': face_id1,
    'second_face_id': face_id2
}

response_similarity = requests.get(url_similarity, params=params_similarity, auth=HTTPBasicAuth(api_key, api_secret))
```
